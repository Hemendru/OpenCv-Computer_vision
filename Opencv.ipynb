```python
import cv2
import numpy as np
img1=cv2.imread("Alluarjun.png")
img2=cv2.imread("prabhas.png")
img1=cv2.resize(img1,(300,400))
img2=cv2.resize(img2,(300,400))
new_img=cv2.addWeighted(img1,1.5,img2,1.1,1)
new=cv2.subtract(img1,img2)


h=np.hstack((new_img,new))
cv2.imshow("Heavy",h)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
img1=cv2.imread("Alluarjun.png")
img2=cv2.imread("prabhas.png")
img1=cv2.resize(img1,(300,400))
img2=cv2.resize(img2,(300,400))
bit_and=cv2.bitwise_and(img1,img2)
bit_or=cv2.bitwise_or(img1,img2)
bit_xor=cv2.bitwise_xor(img1,img2)
bit_not=cv2.bitwise_not(img2)
                       
h=np.hstack((bit_and,bit_or,bit_xor,bit_not))
cv2.imshow("heavy9",h)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img=cv2.imread("Alluarjun.png")
img=cv2.resize(img,(250,250))
img1=cv2.flip(img,0)
img2 = cv2.flip(img,1)
img3 = cv2.flip(img,-1)
h=np.hstack((img1,img2,img3))
img4=cv2.rotate(img,cv2.ROTATE_180)
img5=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
img6=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
h1=np.hstack((img4,img5,img6))
v=np.vstack((h1,h))
cv2.imshow("H1",v)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap=cv2.VideoCapture(0)
def wscube(x):
    pass
cv2.namedWindow("demo")
cv2.createTrackbar("lb","demo",0,255,wscube)
cv2.createTrackbar("lg","demo",0,255,wscube)
cv2.createTrackbar("lr","demo",0,255,wscube)
cv2.createTrackbar("ub","demo",0,255,wscube)
cv2.createTrackbar("ug","demo",0,255,wscube)
cv2.createTrackbar("ur","demo",0,255,wscube)
while cap.isOpened():
    r,frame=cap.read()

    if r==True:
        img=cv2.resize(frame,(400,300))
        hsv_img=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

        lb=cv2.getTrackbarPos("lb","demo")
        lg=cv2.getTrackbarPos("lg","demo")
        lr=cv2.getTrackbarPos("lr","demo")
        ub=cv2.getTrackbarPos("ub","demo")
        ug=cv2.getTrackbarPos("ug","demo")
        ur=cv2.getTrackbarPos("ur","demo")
        lo=np.array([lb,lg,lr])
        up=np.array([ub,ug,ur])

        masks=cv2.inRange(hsv_img,lo,up)
        res=cv2.bitwise_and(img,img,mask = masks)
        
        cv2.imshow("hsv",hsv_img)
        cv2.imshow("wscube",img)
        cv2.imshow("res",res)
        cv2.imshow("masks",masks)
        if cv2.waitKey(1)&0xff==ord("p"):
            break
        
    else:
        break
cap.release()
cv2.destroyAllWindows()
        
```
